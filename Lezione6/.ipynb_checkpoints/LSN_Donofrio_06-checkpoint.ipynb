In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

# Read the file into a pandas DataFrame
energy_metro_data = pd.read_csv('energy_metropolis.dat', header=None, delim_whitespace=True, usecols=[2, 3])
heat_capacity_metro_data = pd.read_csv('heat_capacity_metropolis.dat', header=None, delim_whitespace=True, usecols=[2, 3])
magnetization_metro_data = pd.read_csv('magnetization_metropolis.dat', header=None, delim_whitespace=True, usecols=[2, 3])
magn_susceptibility_metro_data = pd.read_csv('magn_susceptibility_metropolis.dat', header=None, delim_whitespace=True, usecols=[2, 3])

energy_gibbs_data = pd.read_csv('energy_gibbs.dat', header=None, delim_whitespace=True, usecols=[2, 3])
heat_capacity_gibbs_data = pd.read_csv('heat_capacity_gibbs.dat', header=None, delim_whitespace=True, usecols=[2, 3])
magnetization_gibbs_data = pd.read_csv('magnetization_gibbs.dat', header=None, delim_whitespace=True, usecols=[2, 3])
magn_susceptibility_gibbs_data = pd.read_csv('magn_susceptibility_gibbs.dat', header=None, delim_whitespace=True, usecols=[2, 3])

# Extract the values and errors as separate Series
#Final value for all temperature is the twentieth
energy_metro = energy_metro_data[2][19::20].tolist()
energy_metro_err = energy_metro_data[3][19::20].tolist()
energy_gibbs = energy_gibbs_data[2][19::20].tolist()
energy_gibbs_err = energy_gibbs_data[3][19::20].tolist()

heat_capacity_metro = heat_capacity_metro_data[2][19::20].tolist()
heat_capacity_metro_err = heat_capacity_metro_data[3][19::20].tolist()
heat_capacity_gibbs = heat_capacity_gibbs_data[2][19::20].tolist()
heat_capacity_gibbs_err = heat_capacity_gibbs_data[3][19::20].tolist()

magnetization_metro = magnetization_metro_data[2][19::20].tolist()
magnetization_metro_err = magnetization_metro_data[3][19::20].tolist()
magnetization_gibbs = magnetization_gibbs_data[2][19::20].tolist()
magnetization_gibbs_err = magnetization_gibbs_data[3][19::20].tolist()

magn_susceptibility_metro = magn_susceptibility_metro_data[2][19::20].tolist()
magn_susceptibility_metro_err = magn_susceptibility_metro_data[3][19::20].tolist()
magn_susceptibility_gibbs = magn_susceptibility_gibbs_data[2][19::20].tolist()
magn_susceptibility_gibbs_err = magn_susceptibility_gibbs_data[3][19::20].tolist()
```

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
points=16
T = np.linspace(0.5,2.0,num=points)

beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.plot(T, e, label='Analytical')
plt.errorbar(T, energy_metro, yerr = energy_metro_err, fmt = '.', label='Metropolis')
plt.errorbar(T, energy_gibbs, yerr = energy_gibbs_err, fmt = '.', label='Gibbs')
plt.xlabel('T')
plt.ylabel('U/N')
plt.title('Ising 1D, internal energy')
plt.legend()
plt.grid(True)
plt.show()

diff_metropolis = ( e - energy_metro)
diff_gibbs = ( e - energy_gibbs)

plt.errorbar(T, diff_metropolis, yerr = energy_metro_err, fmt = '.', color = 'orange',label='Metropolis')
plt.errorbar(T, diff_gibbs, yerr = energy_gibbs_err, fmt = '.', color = 'g', label='Gibbs')
plt.xlabel('T')
plt.ylabel('Analytical - Simulated (U/N)')
plt.title('Ising 1D, difference between analytical and simulated internal energy')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()
```

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label='Analytical')
plt.errorbar(T, heat_capacity_metro, yerr = heat_capacity_metro_err, fmt = '.', label='Metropolis')
plt.errorbar(T, heat_capacity_gibbs, yerr = heat_capacity_gibbs_err, fmt = '.', label='Gibbs')
plt.xlabel('T')
plt.ylabel('C')
plt.title('Ising 1D, Heat capacity')
plt.legend()
plt.grid(True)
plt.show()

diff_metropolis = (heat - heat_capacity_metro)
diff_gibbs = (heat - heat_capacity_gibbs)

plt.errorbar(T, diff_metropolis, yerr = heat_capacity_metro_err, fmt = '.', color = 'orange', label='Metropolis')
plt.errorbar(T, diff_gibbs, yerr = heat_capacity_gibbs_err, fmt = '.', color = 'g', label='Gibbs')
plt.xlabel('T')
plt.ylabel('Analytical - Simulated (C)')
plt.title('Ising 1D, difference between analytical and simulated heat capacity')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()
```

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

plt.plot(T, M, label='Analytical')
plt.errorbar(T, magnetization_metro, yerr = magnetization_metro_err, fmt = '.', label='Metropolis')
plt.errorbar(T, magnetization_gibbs, yerr = magnetization_gibbs_err, fmt = '.', label='Gibbs')
plt.xlabel('T')
plt.ylabel('M')
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.legend()
plt.grid(True)
plt.show()

diff_metropolis = (M - magnetization_metro)
diff_gibbs = (M - magnetization_gibbs)

plt.errorbar(T, diff_metropolis, yerr = magnetization_metro_err, fmt = '.', color = 'orange', label='Metropolis')
plt.errorbar(T, diff_gibbs, yerr = magnetization_gibbs_err, fmt = '.', color = 'g', label='Gibbs')
plt.xlabel('T')
plt.ylabel('Analytical - Simulated (M)')
plt.title('Ising 1D, difference between analytical and simulated magnetization (h = 0.02)')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()
```

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, label='Analytical')
plt.errorbar(T, magn_susceptibility_metro, yerr = magn_susceptibility_metro_err, fmt = '.', label='Metropolis')
plt.errorbar(T,  magn_susceptibility_gibbs, yerr = magn_susceptibility_gibbs_err, fmt = '.', label='Gibbs')
plt.xlabel('T')
plt.ylabel('X')
plt.title('Ising 1D, Susceptibility')
plt.legend()
plt.grid(True)
plt.show()

diff_metropolis = ( X - magn_susceptibility_metro)
diff_gibbs = ( X - magn_susceptibility_gibbs)

plt.errorbar(T, diff_metropolis, yerr = magn_susceptibility_metro_err, fmt = '.', color = 'orange', label='Metropolis')
plt.errorbar(T, diff_gibbs, yerr = magn_susceptibility_gibbs_err, fmt = '.', color = 'g', label='Gibbs')
plt.xlabel('T')
plt.ylabel('Analytical - Simulated (X)')
plt.title('Ising 1D, difference between analytical and simulated susceptibility')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()
```
