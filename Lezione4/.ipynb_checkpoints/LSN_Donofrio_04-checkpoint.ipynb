### Solid phase

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

#Initial temperature 1.5435
#Step to reach equilibrium 50000

# Read the file into a pandas DataFrame
ekin_data = pd.read_csv('output_ekin_solid.dat', header=None, delim_whitespace=True, usecols=[2, 3])
epot_data = pd.read_csv('output_epot_solid.dat', header=None, delim_whitespace=True, usecols=[2, 3])
etot_data = pd.read_csv('output_etot_solid.dat', header=None, delim_whitespace=True, usecols=[2, 3])
temp_data = pd.read_csv('output_temp_solid.dat', header=None, delim_whitespace=True, usecols=[2, 3])
pres_data = pd.read_csv('output_pres_solid.dat', header=None, delim_whitespace=True, usecols=[2, 3])

# Extract the values and errors as separate Series
ekin = ekin_data[2].tolist()
ekin_err = ekin_data[3].tolist()

epot = epot_data[2].tolist()
epot_err = epot_data[3].tolist()

etot = etot_data[2].tolist()
etot_err = etot_data[3].tolist()

temp = temp_data[2].tolist()
temp_err = temp_data[3].tolist()

pres = pres_data[2].tolist()
pres_err = pres_data[3].tolist()


M = 2000   # Number of throws in each block
N = 20     # Number of blocks

x = np.arange(1, len(ekin_data)+1) * M # Number of throws = block * (Number of throws in each block)

# Plot for Temperature
plt.errorbar(x, temp, yerr=temp_err)
plt.axhline(y=0.8, color='r', linestyle='--', label='Target Temperature')
plt.xlabel('#throws')
plt.ylabel('Temperature')
plt.title('Temperature')
plt.grid(True)
plt.show()

# Plot for Kinetic Energy
plt.errorbar(x, ekin, yerr=ekin_err)
plt.xlabel('#throws')
plt.ylabel('Kinetic Energy')
plt.title('Kinetic Energy')
plt.grid(True)
plt.show()

# Plot for Potential Energy
plt.errorbar(x, epot, yerr=epot_err)
plt.xlabel('#throws')
plt.ylabel('Potential Energy')
plt.title('Potential Energy')
plt.grid(True)
plt.show()

# Plot for Total Energy
plt.errorbar(x, etot, yerr=etot_err)
plt.xlabel('#throws')
plt.ylabel('Total Energy')
plt.title('Total Energy')
plt.grid(True)
plt.show()

# Plot for Pressure
plt.errorbar(x, pres, yerr=pres_err)
plt.xlabel('#throws')
plt.ylabel('Pressure')
plt.title('Pressure')
plt.grid(True)
plt.show()
```

### Comments
Starting from a perfect crystalline configuration, $T$ will not conserve its value.

This behavior is due to the Verlet algorithm, which conserves the total energy and not temperature.
In a perfect crystalline configuration the potential energy is different than the average value at the equilibrium configuration. Consequently, during the initial time evolution, the system undergoes changes in its configuration, leading to variations in both potential and kinetic energy ultimately resulting in variations in the temperature.

To study our system at a specific temperature, we need to explore and determine the initial value of $T$ that would stabilize the average temperature ($T^\star$) after a certain number of time steps (equilibration time). This equilibration time allows the system to reach a state where the average temperature matches the desired value.

In the simulation of the solid and liquid phase we used 20 blocks of lenght 2000. However, in the case of the gas phase, we noticed that the measured values exhibited long fluctuations. As a result, we made the decision to use longer blocks with a length of 4000 to capture the dynamics of the gas phase more effectively. Our choice regarding the appropriate block size for each phase were further confirmed through the analysis of uncertainty estimation with varying block sizes and autocorrelation done in experience 7.
