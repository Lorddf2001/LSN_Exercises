### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
    Remember that given an asset price $S\sim GBM(r,\sigma^2)$, with initial value $S(0)$, then
    $$S(t) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    where $W(t)\sim\mathcal{N}(0,t)$.
    Moreover, since the increments of $W$ are independent and normally distributed, the simple recursive procedure for sampling discretized values of $S\sim GBM(r,\sigma^2)$ at $t_1 < t_2 < â€¦ < t_n$ is:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
</p>

### Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
import numpy
from numpy import exp
from numpy.lib.scimath import log, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

### Numeric solution:

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

t_steps = 100;   # time steps

# Read the file into a pandas DataFrame
Ct_data = pd.read_csv('meanCt.dat', header=None, sep='  ', engine='python')
Pt_data = pd.read_csv('meanPt.dat', header=None, sep='  ', engine='python')
discreteCt_data = pd.read_csv('discretemeanCt.dat', header=None, sep='  ', engine='python')
discretePt_data = pd.read_csv('discretemeanPt.dat', header=None, sep='  ', engine='python')

# Extract the values and errors as separate Series
meanCt = Ct_data[0].tolist()
meanCt_err = Ct_data[1].tolist()

meanPt = Pt_data[0].tolist()
meanPt_err = Pt_data[1].tolist()

discretemeanCt = discreteCt_data[0].tolist()
discretemeanCt_err = discreteCt_data[1].tolist()

discretemeanPt = discretePt_data[0].tolist()
discretemeanPt_err = discretePt_data[1].tolist()


M = 100000   # Total number of throws
N = 100     # Number of blocks
L = int(M/N) # Number of throws in each block

x = np.arange(1, len(meanCt)+1) * L # Number of throws = block * (Number of throws in each block)

# Plot for Mean Call prime
plt.errorbar(x, meanCt, yerr=meanCt_err)
plt.xlabel('#throws')
plt.ylabel('< Call price >')
plt.title('Call price')
plt.grid(True)
plt.axhline(y=call, color='r', linestyle='--') # value for Call price
plt.show()

# Plot for Mean Put price
plt.errorbar(x, meanPt, yerr=meanPt_err)
plt.xlabel('#throws')
plt.ylabel('< Put price >')
plt.title('Put price')
plt.grid(True)
plt.axhline(y=put, color='r', linestyle='--') # value for Put price
plt.show()

# Plot for discrete Mean Call prime
plt.errorbar(x, discretemeanCt, yerr=discretemeanCt_err)
plt.xlabel('#throws')
plt.ylabel('< Call price >')
plt.title(f'Call price (using {t_steps} time intervals)')
plt.grid(True)
plt.axhline(y=call, color='r', linestyle='--') # value for Call price
plt.show()

# Plot for discrete Mean Put price
plt.errorbar(x, discretemeanPt, yerr=discretemeanPt_err)
plt.xlabel('#throws')
plt.ylabel('< Put price >')
plt.title(f'Put price (using {t_steps} time intervals)')
plt.grid(True)
plt.axhline(y=put, color='r', linestyle='--') # value for Put price
plt.show()
```

### Comments

The simulated put and call prices, both direct and discretized, align well with the analytical Black-Scholes solutions.

In the graph of the direct put price we can notice that the simulated value deviates from the exact value by approximately 1.5 times the standard deviation. This discrepancy can be attributed to statistical effects inherent in the simulation process. By changing the random seed used in the simulation, it is possible to observe variations in the simulated value, which can bring it closer to the exact value.

Indeed, the inherent randomness in the simulation leads to statistical fluctuations. These fluctuations are expected and are influenced by the initial conditions and random processes involved in the simulation. By altering the seed, we effectively introduce a different set of random numbers, potentially reducing or eliminating the observed difference.

It's important to note that while changing the seed may help resolve the discrepancy in the specific simulation, the underlying statistical effects will still remain. Therefore, it's necessary to consider the impact of statistical fluctuations and evaluate the overall accuracy and reliability of the simulation results.

All considered, a deviation of 1.5 times the standard deviation is a consistent result.
