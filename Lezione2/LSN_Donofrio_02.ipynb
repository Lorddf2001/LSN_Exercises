### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

# Read the file into a pandas DataFrame
integral_data = pd.read_csv('Integral.dat', header=None, sep='  ', engine='python')
integral_sampling_data = pd.read_csv('Integralsampling.dat', header=None, sep='  ', engine='python')

# Extract the values and errors as separate Series
integral = integral_data[0].tolist()
integral_err = integral_data[1].tolist()

integral_sampling = integral_sampling_data[0].tolist()
integral_sampling_err = integral_sampling_data[1].tolist()

M = 10000   # Total number of throws
N = 100     # Number of blocks
L = int(M/N) # Number of throws in each block

x = np.arange(1, len(integral)+1) * L # Number of throws = block * (Number of throws in each block)

# Plot for Integral
plt.errorbar(x, [val - 1 for val in integral], yerr=integral_err)
plt.xlabel('#throws')
plt.ylabel('<Integral>-1')
plt.title('Integral')
plt.grid(True)
plt.show()

# Plot for Integral Sampling
plt.errorbar(x, [val - 1 for val in integral_sampling], yerr=integral_sampling_err)
plt.xlabel('#throws')
plt.ylabel('<Integral>-1')
plt.title('Integral Sampling')
plt.grid(True)
plt.show()
```

### Comments
As expected, using the importance sampling technique leads to a lower uncertainty for the same amount of throws.

### Exercise 02.2
- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Suggestion: divide your $M$ throws into a reasonable number of blocks, compute $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the step $i\in [0,10^2]$ in each block and use these values (for fixed $i$) to obtain the average value and its statistical uncertainty.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

```python
from scipy.optimize import curve_fit

steplength = 1

# Read the file into a pandas DataFrame
discreteRW_data = pd.read_csv('discretemeanRW.dat', header=None, sep='  ', engine='python')
continuousRW_data = pd.read_csv('continuousmeanRW.dat', header=None, sep='  ', engine='python')

# Replace -nan with 0 in the first error column
discreteRW_data[2].replace(-np.nan, 0, inplace=True)
continuousRW_data[2].replace(-np.nan, 0, inplace=True)

# Extract the values and errors as separate Series
dRWnorm = discreteRW_data[1].tolist()
dRWnorm_err = discreteRW_data[2].astype(float).tolist()

cRWnorm = continuousRW_data[1].tolist()
cRWnorm_err = continuousRW_data[2].astype(float).tolist()

# Determine the number of Random walks
N = len(dRWnorm)
x = np.arange(1, N+1)

# Fitting function
def fit_function(x, steplength):
    return  steplength * np.sqrt(x)

# Discrete RW plot
plt.errorbar(x, dRWnorm, yerr = dRWnorm_err, label='Discrete RW')
plt.xlabel('Number of steps')
plt.ylabel('<norm>')
plt.title('Discrete Random Walk')
plt.grid(True)

# Fit for discrete RW
plt.plot(x, [fit_function(i, steplength) for i in x], label=f'Fit (steplength (={steplength}) * $\sqrt{{N}}$)')
plt.legend()
plt.show()

#Difference between analytical and simulated walk, discrete RW
diff_dRW = ( fit_function(x, steplength) - dRWnorm)
plt.errorbar(x, diff_dRW, yerr = dRWnorm_err, fmt = '.')
plt.xlabel('Number of steps')
plt.ylabel('Analytical - Simulated')
plt.title('Discrete random walk, difference between analytical and simulated walk')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()

# Continuous RW plot
plt.errorbar(x, cRWnorm, yerr = cRWnorm_err, label='Continuous RW')
plt.xlabel('Number of steps')
plt.ylabel('<norm>')
plt.title('Continuous Random Walk')
plt.grid(True)

# Fit for continuous RW
plt.plot(x, [fit_function(i, steplength) for i in x], label=f'Fit (steplength (={steplength}) * $\sqrt{{N}}$)')
plt.legend()
plt.show()

#Difference between analytical and simulated walk, continous RW
diff_cRW = ( fit_function(x, steplength) - cRWnorm)
plt.errorbar(x, diff_cRW, yerr = cRWnorm_err, fmt = '.')
plt.xlabel('Number of steps')
plt.ylabel('Analytical - Simulated')
plt.title('Continuous random walk, difference between analytical and simulated walk')
plt.axhline(y=0.0, color='r', linestyle='--', label='Perfect agreement')
plt.legend()
plt.grid(True)
plt.show()
```

### Comments
The function $f(N)=a\sqrt{N}$ provides a satisfactory fit for both the discrete and continuum random walks. As the value of $N$ increases, the standard deviation $\sigma$ also increases, indicating diffusive behavior.
